```python
from random import seed
from random import random

PassangerId = []
survived = []
seed(1)


for i in range(len(dfTest.index)):
    
    if dfTest.at[i,"Sex"] == 'female':
                
        if dfTest.at[i,"Pclass"] == 2:
            if random() > (1-0.92):
                PassangerId.append(dfTest.at[i,"PassengerId"])
                survived.append(1)
            else:
                PassangerId.append(dfTest.at[i,"PassengerId"])
                survived.append(0)
        
        elif dfTest.at[i,"Pclass"] == 1:
            
            if random() > (1-0.968):
                PassangerId.append(dfTest.at[i,"PassengerId"])
                survived.append(1)
            else:
                PassangerId.append(dfTest.at[i,"PassengerId"])
                survived.append(0)
                
    elif dfTest.at[i,"Sex"] == 'male' and dfTest.at[i,'Parch'] == 0:
        
            if dfTest.at[i,"Pclass"] == 1:
                if random() > (1-0.36):
                    PassangerId.append(dfTest.at[i,"PassengerId"])
                    survived.append(1)
                else:
                    PassangerId.append(dfTest.at[i,"PassengerId"])
                    survived.append(0)

            elif dfTest.at[i,"Pclass"] == 2:
                if random() > (1 - 0.09):
                    PassangerId.append(dfTest.at[i,"PassengerId"])
                    survived.append(1)
                else:
                    PassangerId.append(dfTest.at[i,"PassengerId"])
                    survived.append(0)

            else:
                if random() > (1-0.12):
                    PassangerId.append(dfTest.at[i,"PassengerId"])
                    survived.append(1)
                else:
                    PassangerId.append(dfTest.at[i,"PassengerId"])
                    survived.append(0)
                    
                
```
